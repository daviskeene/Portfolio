```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
words = []
definitions = []

with open('words.txt') as data:
    for line in data:
        if line != "\n":
            # Get rid of new line after line
            line = line.strip('\n')
            # Get rid of anything in quotes
            first_paren = line.find('(')
            dash = line.rfind('-')
            if first_paren != -1:
                if first_paren < dash:
                    line = line[:first_paren] + line[dash:]
                else:
                    pass
            print(line)
            print(line.split(' - '), len(line.split(' - ')))
            word, definition = line.split(' - ')
            words.append(word)
            definitions.append(definition)
            
            
```

```python
df = pd.DataFrame()
```

```python
df['word'] = words
df['definition'] = definitions
```

```python
# Plot a histogram of the first character of the random words
def count_first_letters(words):
    first_letters = [x[0] for x in words]
    d = {}
    for letter in first_letters:
        if letter not in d:
            d[letter] = 1
        else:
            d[letter] += 1
    return d

dictionary = count_first_letters(words)
dictionary = {k: v for k,v in sorted(dictionary.items(), key=lambda x: -x[1])}
```

```python
def count_num_letters(words):
    num_letters = [len(x) for x in words]
    d = {}
    for letter in num_letters:
        if letter not in d:
            d[letter] = 1
        else:
            d[letter] += 1
    return d
```
